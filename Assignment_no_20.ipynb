```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import joblib
```

```python
# Q1: Preparing Features
# Load the Ford Car Dataset
df = pd.read_csv('/content/ford_car_dataset.csv')
df
```

```python
# Independent and Dependent variable
X = df.drop('price', axis=1)
Y = df['price']

print("Independent Features (X):")
print(X)

print("\nDependent Feature (Y):")
print(Y)

# Display Shape
print("\nShape of X:", X.shape)
print("Shape of Y:", Y.shape)
```

```python
# Q2: One-Hot Encoding

# Identify categorical columns
print("Categorical Columns:")
print(X.select_dtypes(include=["object"]).columns)

# Perform One-Hot Encoding
X = pd.get_dummies(X)

# Convert encoded data into integer type
X = X.astype(int)

# Display first 5 rows
print("\n First 5 rows after one- hot encoding:")
print(X.head())
```

```python
# Q3: Feature Scaling

from sklearn.preprocessing import StandardScaler

# Create Standard Scaler
scaler = StandardScaler()

# Numerical Columns
num_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']

# Apply Standard Scaling
X[num_cols] = scaler.fit_transform(X[num_cols])

# Display first 5 rows
print("Scaled Features:")
print(X.head())
```

```python
# Q4: Train-Test Split

from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    Y,
    test_size=0.33,
    random_state=42
)

# Display Shapes
print("Shape of X_train:", X_train.shape)
print("Shape of X_test :", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test :", y_test.shape)
```

```python
# Q5: Building Linear Regression Model

from sklearn.linear_model import LinearRegression

# Create Model
model = LinearRegression()

# Fit Model
model.fit(X_train, y_train)

# print intercept
print("Intercept:")
print(model.intercept_)

# Print Coefficients
coefficients = pd.DataFrame({
    "Feature": X_train.columns,
    "Coefficient": model.coef_
    })
print("\nCoefficients:")
print(coefficients)
```

```python
# Q6: Making Predictions

# Predict on Test Data
y_pred = model.predict(X_test)

print("First 10 Predicted Prices:")
print(y_pred[:10])

# Compare Actual vs Predicted values
comparison = pd.DataFrame({
    "Actual Price": y_test.iloc[:10].values,
    "Predicted Price": y_pred[:10]
})

print("\nFirst 10 Actual vs Predicted Values:")
print(comparison)
```

```python
# Q7: Model Evaluation using R² Score

from sklearn.metrics import r2_score

# Calculate R² Score
r2 = r2_score(y_test, y_pred)

# Print R² Score
print("R² Score:", r2)

# Interpretation
if r2 >= 0.90:
    print("Interpretation: The model fits the data well.")
    print("Comment: Excellent performance.")
elif r2 >= 0.75:
    print("Interpretation: The model fits the data well.")
    print("Comment: Good performance.")
elif r2 >= 0.50:
    print("Interpretation: The model fits the data reasonably well.")
    print("Comment: Average performance.")
else:
    print("Interpretation: The model needs improvement.")
    print("Comment: Poor performance.")

# Interpretation:
# The R² score shows how well the Linear Regression model predicts the car price.
# A value close to 1 means the model performs very well.
# A value close to 0 means the model performs poorly.
```

```python
# Q8: Saving the Model
import joblib

# Save Linear Regression Model
joblib.dump(model, "LR_ford_car.pkl")

# Save StandardScaler
joblib.dump(scaler, "scaler.pkl")

# Save Column Names
joblib.dump(X.columns.tolist(), "columns.pkl")

print("Linear Regression model created and saved successfully.")
```
