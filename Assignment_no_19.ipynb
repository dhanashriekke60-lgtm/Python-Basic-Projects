```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Q1: Load House Price Prediction Dataset
filepath = '/content/House_Price_Prediction.csv'
df = pd.read_csv(filepath)
df
```

```python
# Q2: Dataset Selection & Initial Exploration

# Display First 10 rows
print("Display First 10 rows:")
df.head(10)

# Observation:
# I selected the Housing dataset because it contains house prices and different property features.
# The target variable is price, which is continuous.
# This dataset is useful for regression problems.
# It contains both numerical and categorical columns.
# It helps understand how different features affect house prices.
# The goal is to predict the house price.
```

```python
# Q3: Missing & Duplicate Values Analysis
# Missing values
print("Missing Value:")
print(df.isnull().sum())

# Missing value percentage
print("\nMissing value percentage:")
print((df.isnull().sum()/len(df))*100)

# Remove duplicates
df = df.drop_duplicates()

print("\nDuplicate Rows:")
print(df.duplicated().sum())

# Observations:
# No missing values (if output is 0).
# Duplicate rows removed successfully.
# Dataset is clean for analysis.
```

```python
# Q4: Statistical Summary
print("Statistical Summary:")
print(df.describe())

print("\nMinimum Price :")
print(df["price"].min())

print("\nMaximum Price :")
print(df["price"].max())

print("\nMean Price :")
print(df["price"].mean())

print("\nMedian Price :")
print(df["price"].median())

# Observation:
# Mean is the average price.
# Median is the middle value.
# Maximum price is much higher than minimum.
# Data shows variation in house prices.
```

```python
# Q5: Histogram Analysis of Numeric Features
num_cols = df.select_dtypes(include=["int64","float64"]).columns

plt.figure(figsize=(14,8))

for i,col in enumerate(num_cols):
    plt.subplot(2,3,i+1)
    plt.hist(df[col], bins=20, color="blue", edgecolor="black")
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Q6: Count Plots of Categorical Features
cat_cols = df.select_dtypes(include="object").columns

plt.figure(figsize=(15,12))

for i,col in enumerate(cat_cols):
    plt.subplot(3,3,i+1)
    sns.countplot(data=df,x=col)
    plt.xticks(rotation=45)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Q7: Correlation Heatmap
plt.figure(figsize=(10,7))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap="coolwarm")

plt.show()

# Observation:
# Dark colors indicate stronger correlation.
# Features with higher correlation to price are more useful.
# Weak correlation features have less effect.
```

```python
# Q8: Feature Identification
X = df.drop("price", axis=1)

y = df["price"]

print("Independent Features")
print(X.columns)

print("\nDependent Feature")
print(y.name)

# Justification
# Independent Features (X): These variables describe the characteristics of a house (such as
# area, bedrooms, bathrooms, parking, etc.) and are used as input to predict the house price.
# Dependent Feature (y): The price column is the target variable because it is a continuous
# numerical value that we want to predict using the independent features.
```

```python
# Q9: Encoding Categorical Variables

# Before transformation
print("Before Encoding:")
print(df[["mainroad","guestroom"]].head())

# One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['mainroad','guestroom'])

# After transformation
print("\nAfter Encoding:")
print(df_encoded.head())

# Explanation of Encoding Choices:
# Used One-Hot Encoding for categorical columns.
# Converts text values into 0 and 1.
# Makes the data suitable for machine learning models.
# Does not create any order between categories.
# Improves model performance and accuracy.
```
