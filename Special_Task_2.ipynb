```python
# Bar Chart
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,5))

sns.countplot(
    x='Survived',
    hue='Survived',
    data=df,
    palette=['red', 'blue'],
    legend=False
)

plt.title("Passenger Survival Count", fontsize=16)
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
plt.hist(
    df['Age'].dropna(),
    bins=20,
    color='deepskyblue',
    edgecolor='black'
)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(7,5))
plt.scatter(
    df['Age'],
    df['Fare'],
    color='purple',
    alpha=0.6
)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

```python
# Box Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,5))
sns.boxplot(
    y=df['Fare'],
    color='orange'
)
plt.title("Fare Distribution")
plt.ylabel("Fare")

plt.show()
```

```python
# Heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9,7))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm',
    linewidths=0.5
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Pie Chart
import matplotlib.pyplot as plt
import seaborn as sns

gender = df['Sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(
    gender,
    labels=gender.index,
    autopct='%1.1f%%',
    colors=['skyblue','lightpink'],
    startangle=90,
    shadow=True
)
plt.title("Gender Distribution")
plt.show()
```
